```python
from transformers import pipeline
from youtube_transcript_api import YouTubeTranscriptApi
```

**Enter The link for the video which you want to summarize**



```python
print('Please Enter the link to the YouTube Video')
link = input()
```

```python
video_id = link.split("=")[1]
transcript = YouTubeTranscriptApi.get_transcript(video_id)
full_transcript = ""
for x in transcript:
    full_transcript += " " + x['text']
print(full_transcript)
```

```python
print(len(full_transcript))
```

```python
# using pipeline API for summarization task
summarizar = pipeline("summarization")
```

**Get the Summary of the Video**

```python
itr = int(len(full_transcript)/1000)
summary_text = []
for i in range(0, itr+1):
    start = 0
    start = i*1000
    end = (i+1)*1000
    print("Input text \n" + full_transcript[start:end])
    out = summarizar(full_transcript[start:end],max_length=90)
    out = out[0]
    out = out["summary_text"]
    print("Summarized text\n"+out)
    summary_text.append(out)
```

```python
print(summary_text)
```
